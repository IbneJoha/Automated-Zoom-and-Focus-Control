```python
import serial
import time

# Function to set up serial communication with Arduino
def setup_serial_connection(port, baudrate=9600):
    try:
        arduino = serial.Serial(port, baudrate)  # Open serial port
        time.sleep(0.1)  # Wait for the serial connection to establish
        return arduino
    except serial.SerialException as e:
        print(f"Error: {e}")
        return None

# Function to calculate Focus based on the zoom value (Z) using the fitted 5th-degree polynomial equation
def calculate_focus(Z):
    return (4.7105e-10 * Z**5) - (1.7064e-7 * Z**4) + (1.9530e-5 * Z**3) - (2.0363e-3 * Z**2) - (0.5778 * Z) + 240.7969

# Set up serial communication
arduino = setup_serial_connection('COM6')  # Replace 'COM6' with your Arduino port

if arduino is None:
    print("Failed to open serial port. Exiting.")
    exit()

# Loop through zoom values (from 152 to 248)
for Z in range(80, 244, 1):  # Zoom values from 152 to 248 with step of 1
    try:
        # Calculate the Focus value
        desiredValue1 = calculate_focus(Z)

        # Round the Focus value to the nearest integer
        desiredValue1 = round(desiredValue1)
        desiredValue2 = Z

        # Check if the serial port is still open
        if arduino.is_open:
            # Send the Zoom value first, followed by the Focus value
            arduino.write(f"{desiredValue2} {desiredValue1}\n".encode())
            print(f"Sent Zoom: {desiredValue2} | Focus: {desiredValue1}")
        else:
            print("Serial port closed unexpectedly. Reopening...")
            arduino = setup_serial_connection('COM6')  # Reopen the port if it is closed
            if arduino is None:
                print("Failed to re-open serial port. Exiting.")
                exit()

        # Wait 300ms before sending the next pair of values
        time.sleep(0.35)  # 300ms delay

    except Exception as e:
        print(f"Error during serial communication: {e}")

# Close the serial port to release it
arduino.close()
print("Serial port closed.")
```

```python
# Function to set up serial communication
def setup_serial_connection(port, baudrate=9600):
    while True:
        try:
            arduino = serial.Serial(port, baudrate, timeout=1)  # Open serial port
            time.sleep(0.1)  # Allow time for the connection to establish
            print(f"Connected to {port} at {baudrate} baud.")
            return arduino
        except serial.SerialException as e:
            print(f"Error: {e} | Retrying in 3 seconds...")
            time.sleep(3)  # Retry after 3 seconds if the connection fails

# Function to calculate Focus based on the zoom value (Z) using the 5th-degree polynomial equation
def calculate_focus(Z):
    return (4.7105e-10 * Z**5) - (1.7064e-7 * Z**4) + (1.9530e-5 * Z**3) - (2.0363e-3 * Z**2) - (0.5778 * Z) + 240.7969

# Set up serial communication
port_name = "COM6"  # Change this based on your Raspberry Pi serial port
arduino = setup_serial_connection(port_name)

# Iterate through zoom values from 80 to 244
for Z in range(80, 244):  # Range from 80 to 244
    try:
        desiredValue1 = round(calculate_focus(Z))  # Round focus value
        desiredValue2 = Z

        if arduino.is_open:
            data_to_send = f"{desiredValue2} {desiredValue1}\n"
            arduino.write(data_to_send.encode())  # Send encoded data
            print(f"Sent -> Zoom: {desiredValue2} | Focus: {desiredValue1}")
        else:
            print("Serial port closed unexpectedly. Attempting to reconnect...")
            arduino = setup_serial_connection(port_name)

        time.sleep(0.35)  # 350ms delay

    except serial.SerialException as e:
        print(f"Serial communication error: {e}. Reconnecting...")
        arduino = setup_serial_connection(port_name)

    except Exception as e:
        print(f"Unexpected error: {e}")

# Once last value (248) is reached, keep sending it continuously
print("Reached last Zoom value. Continuously sending Zoom: 248 and calculated Focus.")

while True:
    try:
        desiredValue1 = round(calculate_focus(248))
        desiredValue2 = 248

        if not arduino.is_open:
            print("Serial port closed unexpectedly. Attempting to reconnect...")
            arduino = setup_serial_connection(port_name)

        # Flush the output buffer before sending new data (prevents data buildup)
        arduino.reset_output_buffer()

        data_to_send = f"{desiredValue2} {desiredValue1}\n"
        arduino.write(data_to_send.encode())  # Send encoded data
        print(f"Continuously Sent -> Zoom: {desiredValue2} | Focus: {desiredValue1}")

        time.sleep(0.4)  # Maintain a steady interval for continuous transmission

    except serial.SerialException as e:
        print(f"Serial communication error: {e}. Reconnecting...")
        arduino = setup_serial_connection(port_name)

    except Exception as e:
        print(f"Unexpected error: {e}")
```
