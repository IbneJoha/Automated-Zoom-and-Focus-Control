```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Given updated data
zoom_values = np.array([8, 12, 16, 20, 24, 28, 32, 36, 40, 44, 48, 52, 56, 60, 64, 68, 72, 76, 80, 84, 88, 92, 96, 100, 104, 108, 112, 116, 120, 124,128, 132, 
                        136, 140, 144, 148, 152, 156, 160, 164, 168, 172, 176, 180, 184, 188, 192, 196, 200, 204, 208, 212, 216, 220, 224, 228, 232, 236, 240, 244])
focus_values = np.array([236, 234, 232, 228, 226, 222, 220, 218, 216, 212, 210, 207, 204, 201, 200, 196, 192, 190, 186, 183, 180, 176, 172, 169, 166, 162, 160, 156, 152, 148, 146, 141,
                         138, 133, 130, 125, 122, 117, 114, 110, 108, 102, 98, 94, 90, 88, 84, 81, 78, 74,  72, 70, 68, 66, 64, 64, 64, 64, 64, 64])

# Polynomial Fit (4th-degree)
poly_coeffs = np.polyfit(zoom_values, focus_values, 4)  # Updated to 4th degree
poly_eqn = np.poly1d(poly_coeffs)

# Exponential Fit Function
def exp_func(x, a, b, c):
    return a * np.exp(b * x) + c

# Fit exponential curve
exp_params, _ = curve_fit(exp_func, zoom_values, focus_values, p0=(200, -0.01, 60))

# Generate fitted values
zoom_fit = np.linspace(min(zoom_values), max(zoom_values), 100)
focus_poly_fit = poly_eqn(zoom_fit)
focus_exp_fit = exp_func(zoom_fit, *exp_params)

# Plot the data and the polynomial & exponential fits
plt.scatter(zoom_values, focus_values, label="Given Data", color='red')
plt.plot(zoom_fit, focus_poly_fit, label="4th-Degree Polynomial Fit", color='blue')  # Updated label
plt.plot(zoom_fit, focus_exp_fit, label="Exponential Fit", color='green', linestyle='dashed')
plt.xlabel("Zoom Value")
plt.ylabel("Focus Value")
plt.legend()
plt.title("Zoom vs Focus Relationship - Improved Fit")
plt.grid(True)
plt.show()

# Display the polynomial and exponential equations
print("Polynomial Equation:")
print(poly_eqn)
print("\nExponential Equation: f(x) = {:.4f} * exp({:.4f} * x) + {:.4f}".format(*exp_params))
```
